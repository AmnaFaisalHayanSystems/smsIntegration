```python
# A script that's needed to setup django if it's not already running on a server.
# Without this, you won't be able to import django modules
import sys, os, django
from django.conf import settings

# Find the project base directory
BASE_DIR = settings.BASE_DIR

# Add the project base directory to the sys.path
# This means the script will look in the base directory for any module imports
# Therefore you'll be able to import analysis.models etc
sys.path.insert(0, BASE_DIR)

# The DJANGO_SETTINGS_MODULE has to be set to allow us to access django imports
os.environ.setdefault("DJANGO_SETTINGS_MODULE", "eq.settings")

#  Allow queryset filtering asynchronously when running in a Jupyter notebook
os.environ["DJANGO_ALLOW_ASYNC_UNSAFE"] = "true"

# This is for setting up django
django.setup()
```

```python
from datasource.models import formatedData
import json

def create_or_update_data(i):
    data = json.loads(i['user_details'])
    
    try:
        obj, created = formatedData.objects.get_or_create(
             name = data['first_name'] + " " + data['last_name'],
             cnic = data['cnic'],
             contactNumber = data['contact_number'] if "contact_number" in data else data['contactNumber'],
             email = data['email'],
             data = data,
             address = data['address'],
        )
        if created:
            print("Created new entry: " + obj.name)
        else:
            print('Exists: ' + str(obj.id) + " - " + obj.name)
    except Exception as e:
        print(data)
        print(e)
    
    try:
        obj, created = formatedData.objects.get_or_create(
             name = data['beneficiary_name'],
             cnic = data['beneficiary_cnic'],
             contactNumber = data['beneficiary_contact_number'],
             email = data['beneficiary_email'],
             data = data,
             address = data['address'],
        )
    except Exception as e:
        print(data)
        print(e)
    
    
    try:
        obj, created = formatedData.objects.get_or_create(
             name = data['spouse_first_name'] + " " + data['spouse_last_name'],
             cnic = data['spouse_cnic'],
             contactNumber = data['contact_number'] if "contact_number" in data else data['contactNumber'],
             email = data['email'],
             data = data,
             address = data['address'],
        )
    except Exception as e:
        print(data)
        print(e)
```

```python
  
# Opening JSON file
f = open('../data/order-2022-06-21.json')
  
# returns JSON object as 
# a dictionary
data = json.load(f)
  
# Iterating through the json
# list
for i in data:
    create_or_update_data(i)
  
# Closing file
f.close()
```
